```python
import tweepy
from textblob import TextBlob
```

```python
def get_tweet_sentiment(tweet_text):
    # Perform sentiment analysis on tweet text using TextBlob
    blob = TextBlob(tweet_text)
    polarity = blob.sentiment.polarity

    # Assign sentiment based on polarity value
    if polarity > 0:
        sentiment = 'positive'
    elif polarity < 0:
        sentiment = 'negative'
    else:
        sentiment = 'neutral'

    return sentiment
```

```python
def get_tweet_replies_info(tweet_id, consumer_key, consumer_secret, access_token, access_token_secret):
    # Authenticate with Twitter API
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)

    # Create API object
    api = tweepy.API(auth)

    # Retrieve replies to the tweet
    replies = tweepy.Cursor(api.search_tweets, q=f'to:{tweet_id}', tweet_mode='extended').items()

    # Define counters
    replies_count = 0
    positive_replies_count = 0
    negative_replies_count = 0

    # Analyze each reply
    for reply in replies:
        reply_sentiment = get_tweet_sentiment(reply.full_text)

        replies_count += 1
        if reply_sentiment == 'positive':
            positive_replies_count += 1
        elif reply_sentiment == 'negative':
            negative_replies_count += 1

    return replies_count, positive_replies_count, negative_replies_count
```

```python
def search_tweets(search_query, consumer_key, consumer_secret, access_token, access_token_secret):
    # Authenticate with Twitter API
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)

    # Create API object
    api = tweepy.API(auth)

    # Define counters
    total_direct_mentions = 0
    total_positive_direct_mentions = 0
    total_negative_direct_mentions = 0
    total_indirect_mentions = 0
    total_positive_indirect_mentions = 0
    total_negative_indirect_mentions = 0
    total_retweets = 0
    total_likes = 0
    total_indirect_followers = 0

    # Search for tweets
    for tweet in tweepy.Cursor(api.search_tweets, q=search_query, tweet_mode='extended').items():
        # Increment direct mentions counter
        total_direct_mentions += 1

        # Get tweet information
        likes = tweet.favorite_count
        retweets = tweet.retweet_count
        sentiment = get_tweet_sentiment(tweet.full_text)
        followers_count = tweet.user.followers_count

        # Increment sentiment-based direct mentions counters
        if sentiment == 'positive':
            total_positive_direct_mentions += 1
        elif sentiment == 'negative':
            total_negative_direct_mentions += 1

        # Increment retweets and likes counters
        total_retweets += retweets
        total_likes += likes

        # Increment indirect mentions and followers counters
        replies_count, positive_replies_count, negative_replies_count = get_tweet_replies_info(tweet.id)
        total_indirect_mentions += replies_count
        total_positive_indirect_mentions += positive_replies_count
        total_negative_indirect_mentions += negative_replies_count
        total_indirect_followers += followers_count

    # Print the results
    print("Total Direct Mentions:", total_direct_mentions)
    print("Total Positive Direct Mentions:", total_positive_direct_mentions)
    print("Total Negative Direct Mentions:", total_negative_direct_mentions)
    print("Total Indirect Mentions:", total_indirect_mentions)
    print("Total Positive Indirect Mentions:", total_positive_indirect_mentions)
    print("Total Negative Indirect Mentions:", total_negative_indirect_mentions)
    print("Total Retweets:", total_retweets)
    print("Total Likes:", total_likes)
    print("Total Indirect Followers:", total_indirect_followers)
```
